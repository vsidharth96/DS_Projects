## Domain Analysis (Raw data)
1. Date : Date of invoice
2. Particulars : Customer / Vendor Name
3. Vch Type : sales - sales vouchers ; Purchase - purchase vouchers
4. Vch No : voucher number (Not required)
5. VALUE : Total sales or purchase items values (inclusive of 7% GST)
6. month name : Name of the month (Not required)
7. Year name : Name of the year (Not required)
8. gross Profit : "+" if sales and "-" if purchase (Not required)
9. STV RPF : Name of the two companies
